### 1- Imposto a pagar no Lucro Presumido

- 5% sobre faturamento de ISS (mensal)
- 0,65% de PIS sobre faturamento, (mensal)
- 3% de COFINS sobre faturmaneto, (mensal)
- 4.8% de IR (trimestral)
- 10% de IR Adicional sobre o que ultrapassar 20mil do faturamento (trimestral)
- CSLL: 2,88% sobre faturamento (trimestral)

```python
import numpy as np
faturamento = {
    'jan': 'R$ 95.141,98',
    'fev': 'R$ 95.425,16',
    'mar': 'R$ 89.716,31',
    'abr': 'R$ 78.459,99',
    'mai': 'R$ 71.087,28',
    'jun': 'R$ 83.911,06',
    'jul': 'R$ 56.467,26',
    'ago': 'R$ 88.513,58',
    'set': 'R$ 66.552,49',
    'out': 'R$ 80.164,07',
    'nov': 'R$ 66.964,33',
    'dez': 'R$ 71.525,25',
}

# você precisa inserir no sistema um dicionário no formato:

# resultado = {
#     'mes': (faturamento, imposto_mensal, imposto_trimestral),
# }

iss = 0.05
pis = 0.0065
cofins = 0.03
ir = 0.048
ir_plus = 0.1
csll = 0.0288

def alterar_string(fat):
    for mes in fat:
        fat[mes] = fat[mes].replace('R$ ', '')
        fat[mes] = fat[mes].replace('.', '')
        fat[mes] = fat[mes].replace(',', '.')
    return fat

def convert_to_float(fat):
    for valor in fat:
        fat[valor] = float(fat[valor])
    print(type(fat['jan']))
    return fat


def faturamento_mensal(fat, iss, pis, cofins):
    lista_imposto_mensal = []
    impo_iss = 0
    impo_pis = 0
    impo_cofins = 0
    imposto_mensal = 0
    for valor in fat:
        impo_iss = fat[valor] * iss
        impo_pis = fat[valor] * pis
        impo_cofins = fat[valor] * cofins
        imposto_mensal = fat[valor] - (impo_iss + impo_pis + impo_cofins)
        lista_imposto_mensal.append(imposto_mensal)
    return lista_imposto_mensal

def faturamento_trimestral(fat, ir, ir_plus, csll):
    values_fat = list(fat.values())

    splits = np.array_split(values_fat, 4)
    lista_imposto_trimestral = []

    for lista in splits:
        aux = 0
        impo_ir = 0
        impo_ir_plus = 0
        impo_csll = 0
        imposto_trimestral = 0.0
        lista_imposto_trimestral.append(imposto_trimestral)
        lista_imposto_trimestral.append(imposto_trimestral)
        for i in lista:
            aux += i
        impo_ir = aux * impo_ir
        if aux > 20000:
            impo_ir_plus = aux * ir_plus
        else:
            impo_ir_plus = 0
        impo_csll = aux * csll
        imposto_trimestral = aux - (impo_ir + impo_ir_plus + csll)
        lista_imposto_trimestral.append(imposto_trimestral)
    return lista_imposto_trimestral

def juntar_tudo(fatu, imposto_mensal, imposto_trimestal):
    chaves = list(fatu.keys())
    values = list(fatu.values())
    #print(chaves, '\n', values)
    resultado = []
    for i, fat in enumerate(chaves):
        resultado.append((values[i], imposto_mensal[i],imposto_trimestral[i]))
    
    print(resultado, '\n')
    
    faturamento_anual = dict(zip(chaves, resultado))
    
    return faturamento_anual
   
  
```

```python
fat_string = alterar_string(faturamento)
#print(fat_string, '\n')

fat_to_float = convert_to_float(fat_string)
#print(fat_to_float, '\n')

imposto_mensal = faturamento_mensal(fat_to_float, iss, pis, cofins)
#print(imposto_mensal, '\n')

imposto_trimestral = faturamento_trimestral(fat_to_float, ir, ir_plus, csll)
#print(imposto_trimestral, '\n')

fat_anual = juntar_tudo(fat_to_float, imposto_mensal, imposto_trimestral)
print(fat_anual)
```
